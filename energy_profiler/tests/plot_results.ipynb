```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
```

```python
latency = {'Apple M1 CPU': 5.88, 'Apple M1 GPU': 3.89, 'Nvidia A100': 3.52} # s
energy = {'Apple M1 CPU': 47.09, 'Apple M1 GPU': 7.62, 'Nvidia A100': 134.19} # J
power = {'Apple M1 CPU': 20.8, 'Apple M1 GPU': 6.4, 'Nvidia A100': 42} # W

colors = {'latency': '#4292C6', 'energy': '#F1C40F', 'power': '#E67E22'}
metric_labels = {'latency': 'Latency (s)', 'energy': 'Energy Consumption (J)', 'power': 'Peak Power Draw (W)'}

x = np.arange(len(latency))

width = 0.2
fig, ax = plt.subplots()

count = 0
for metric in metric_labels.keys():
    if metric == 'latency':
        values = latency.values()
    elif metric == 'energy':
        values = energy.values()
    else:
        values = power.values()
    ax.bar(x - 1*width + width*count, values, 
         width, label=metric_labels[metric], color=colors[metric])
    count += 1

ax.set_ylabel('Hardware Perf. Measure (s/J/W)', fontsize=12)
ax.set_xticks(x)
ax.set_xticklabels(latency.keys(), fontsize=12)

ax.legend()

fig.tight_layout()
plt.grid(axis = 'y', linestyle='--')
plt.savefig(f'apple_m1_gains.pdf', bbox_inches = 'tight')
plt.show()
```
